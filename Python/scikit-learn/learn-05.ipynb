### Score vs Accuracy score

#### In `score` we give the Question and Our predicted answer and it internally takes the question and passes it to prediction and returns the predicted data that is actually like the answer sheet and we match it with the passed data that is actually like an answer key 
```
score(X_test, y_test) -> 
predict(X_test) -> y_predict
y_predict == y_test
```

#### In `Accuracy score`.. a matrix function.. we just give the predicted data and ask for it to check if it is predicted correctly or not

```
Accuracy_score(y_predict, y_test)
```

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.01, random_state=0)

clf = LogisticRegression(max_iter=100)
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)

print(accuracy_score(y_pred, y_test))
````
