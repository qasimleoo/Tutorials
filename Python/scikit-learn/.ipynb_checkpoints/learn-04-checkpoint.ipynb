### Train_Test_Split

Data is split in 4 parts (vertically and horizontally)

Rows
Train (70-80%) for training (T)
Testing (30-20%) for testing (t)

Columns
Vertical Split (VS1) .. Dependent variable on Independent vars

````python
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
````

````python
data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=0)

X_train
````

````python
clf = LogisticRegression(max_iter=200)
clf.fit(X_train, y_train)
````

````python
clf.score(X_test, y_test)
````

### Robust Scaler and Outliers
A preprocessing technique

- (Extreme) Outliers actually make our data worst ..  like in a class 98% student get 40 to 60 marks but 2% get 100% marks .. normally average is 50 but 'cause of those 2% .. average goes to 70.. which has negative impact on 98%..


```
x' = (x - m(Q2) ) / IQR
IQR = Q3 - Q1 (Inter quartile range)

x -> nth val in table
m(Q2) -> median


Q0  Q1  Q2  Q3  Q4
0   25  50  75  100

─────
|10 | Q0
|11 | (11 + 12) / 2 
|12 | Q1 = 11.5
|13 | Q2
|14 | (14 + 15) / 2
|15 | Q3 = 14.5
|100| Q4
─────

m = 13 
IQR = 14.5 - 11.5 = 3
x0 = (10 - 13) / 3 = -1 

x1 = (11 - 13) / 3 = -2 / 3

x2 = (12 - 13) / 3 = -1 / 3

x3 = (13 - 13) / 3 = 0

x4 = (14 - 13) / 3 = 1 / 3

x5 = (15 - 13) / 3 = 2 / 3

x6 = (100 - 13) / 3 = 87 / 3

```

````python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import RobustScaler

data = np.array([10,11,12,13,14,15,100])
data = data.reshape(-1,1)
````

````python
rob_scaled = RobustScaler().fit_transform(data)
rob_scaled
````

````python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.scatter(range(len(data)), data, c='red')
plt.title("Original Data")
plt.ylabel("Value")

plt.subplot(1,3,2)
plt.scatter(range(len(rob_scaled)), rob_scaled, c='green')
plt.title("After Robust Scaler")

# plt.show()
````
