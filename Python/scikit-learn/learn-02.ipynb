### MinMax Scaler

#### Formula: X' = (X - Xmin) / (Xmax - Xmin)

```
X -> 1 2 3 4 5

Xmin = 1, Xmax = 5

Take first -> 1
X0 = (1 - 1) / (5 - 1) = 0
X1 = (2 - 1) / (5 - 1) = 1/4 = 0.25
X2 = (3 - 1) / (5 - 1) = 2/4 = 0.50
X3 = (4 - 1) / (5 - 1) = 3/4 = 0.75
X4 = (5 - 1) / (5 - 1) = 4/4 = 1.00
```

````python
from sklearn.preprocessing import MinMaxScaler
````

````python
import numpy as np

df = pd.DataFrame({
    'C1': [1,2,3,4,5],
    'C2': [6,7,8,12,14]
}, index=['a', 'b', 'c', 'd', 'e'])

df
````

### Binarizer Preprocessing - to binarize data


````python
from sklearn.preprocessing import Binarizer
````

````python
b = Binarizer(threshold=5) 
# greater than 5 will be marked as 1 and less than or equal 5 will be marked as 0
# default threshold is 0

# random data can't be handled properly .. so we binarize it first
arr = np.array([[13, 10, 21], [21, 2, 1], [3, 5, 5]]) # 2D array

binary1 = b.transform(arr)
# binary1 = b.fit_transform(arr)
binary1

# used in Bernoulli Naive Bayes
````
