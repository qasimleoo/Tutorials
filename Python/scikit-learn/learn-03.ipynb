## Generated Datasets

you wanna perform regressions .. wanna train data but you don't have the data .. what will you do in that case?


### Make Regression

#### Regression?
Independent Variable vs Dependent Variable
X - Y
X0 X1 X2 X3 X4 - Y (Continuos vals instead of 0s and 1s)

- Classification -> Categorical Data (0, 1, Yes, No, Pass, Fail, True, False)
- Regression -> Continuos Data (Not 0s or 1s) .. but proper data

```python
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_features=1, noise=5, n_samples=5000, random_state=0)
# n_features -> x (Independent Vars)
# noise -> Disturbance in data
# n_samples -> total records
# random_state -> 0 means gimme the same data on each run .. removing it will give you a random data on each load .. both y and x

x 
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, s=1, c='green')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
x,y = make_regression(n_features=4, noise=5, n_samples=1000)
x
```

### Classification Dataset
make_classification

- Classification -> Categorical Data (0, 1, Yes, No, Pass, Fail, True, False)
- Regression -> Continuos Data (Not 0s or 1s) .. but proper data

```python
from sklearn.datasets import make_classification 
```

```python
x, y = make_classification(n_features=10, n_samples=200000, random_state=0, n_classes=10, n_informative=5)

x[:,0]
```

```python
import matplotlib.pyplot as plot
plot.scatter(x[:,0], x[:,2], c=y, s=1)
```
