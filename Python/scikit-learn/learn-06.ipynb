### Handling missing values 

##### Simple Imputer and KNN Imputer

```
Age  Salary
25.0 50000.0
NaN  60000.0
30.0 52000.0
28.0 NaN
NaN  58000.0
```

#### Simple Imputer (Globally .. decided for has to be for each)

Many strategies but main ones are these:
- Mean - takes mean of all values and sets that for each NaN
- Median - sorts the data and takes median and sets it for each NaN
- Constant - if you decide to set a value of 100 it will set that for each


#### KNN Imputer (Locally)
- based on NN
- based on k value
- You wanna know age .. take its salary val.. and get NN of that salary.. suppose k=2 .. now take nearest 2 vals .. 52 and 58 .. but 58 has age NaN so move to next i.e., 50 .. get ages of these salaries and take mean (25+30)/2 = 27
- same goes for next NaN
  

````python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer

data = {
    'Age': [25, np.nan, 30, 28, np.nan],
    'Salary': [50000, 60000, np.nan, 52000, 58000]
}

df = pd.DataFrame(data)
# print("Original Data:\n", df.to_string(index=False))
print("Original Data:\n", df)

simple_imputer = SimpleImputer(strategy='mean')
df_simple = pd.DataFrame(simple_imputer.fit_transform(df), columns=df.columns)
print("\n\nMean Strategy: \n", df_simple)

simple_imputer = SimpleImputer(strategy='median')
df_simple = pd.DataFrame(simple_imputer.fit_transform(df), columns=df.columns)
print("\n\nMedian Strategy: \n", df_simple)

simple_imputer = SimpleImputer(strategy='constant', fill_value=100) # but it will fill at all columns.. 
df_simple = pd.DataFrame(simple_imputer.fit_transform(df), columns=df.columns)
print("\n\nConstant Strategy: \n", df_simple)

simple_imputer = SimpleImputer(strategy='constant', fill_value=50000) # but it will fill at all columns.. 
df_simple = pd.DataFrame(simple_imputer.fit_transform(df[['Salary']]), columns=['Salary'])
print("\n\nConstant Strategy: \n", df_simple)

knn_imputer = KNNImputer(n_neighbors=2)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df), columns=df.columns)
print("\n\nKNN: \n", df_knn)
````
