```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 100000], # Master Biryani
    [2, 120000, 135000, 160000, 130000], # Sheikh G Hotel
    [3, 90000, 110000, 130000, 145000],  # Dogar Restaurant
    [4, 200000, 210000, 240000, 260000], # AA Pakwan
    [5, 170000, 165000, 190000, 205000]  # Burger Point
])

print("======= Sales Analysis =======")
print("\n Shape: ", sales_data.shape)

print("\n Sample data for 1st 3 restaurants: ", sales_data[:3])
# print("\n Sample data for 1st 3 restaurants: ", sales_data[:, 1:])
```

```python
# Total sales per year
# axis = 0 .. cause we wanna get sales of each year
# sales_data[:, 1:]

print("Total for each year: ", np.sum(sales_data[:, 1:], axis=0))
print("Total yearly sale: ", np.sum(sales_data[:, 1:]))
```

```python
# Min sales
# axis = 1 .. cause we wanna get sales of each hotel
print("Min sales of each year of each hotel: ", np.min(sales_data[:, 1:], axis=1))

# print("Sum of min sales: ", np.sum(np.min(sales_data[:, 1:], axis=1)))
```

```python
# Max sales
# axis = 1 .. cause we wanna get sales of each hotel
print("Max sales of each year of each hotel: ", np.max(sales_data[:, 1:], axis=1))

# print("Sum of Max sales: ", np.sum(np.max(sales_data[:, 1:], axis=1)))
```

```python
# Average sales

avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales: ", avg_sales)
```

```python
# Cumulative sales .. pre + now -> pre + now

cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sale: ", cum_sales)

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(sales_data[:, 1:], axis=1).T)
plt.title("Average cumulative Sales per year")
plt.xlabel("Year")
plt.xlabel("Sales")
plt.grid(True)
plt.show()
```

```python
## Vectors

vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Addition: ", vector1 + vector2)

print("Multiplication (element wise): ", vector1 * vector2)

print("Dot product (dot product): ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / np.linalg.norm(vector1) * np.linalg.norm(vector2))
print("ArcCos Angle: ", angle)

# Matrix multiply not possible here as it is not possible for 1D vectors
# Let's try it
# vector1 = np.array([[[1,2,3]]])
vector1 = np.array([[1,2,3], [4,5,6]])
vector2 = np.array([[1,2,3]])
print(vector1.shape, vector2.shape)

print("Matrix multiplication: ", np.dot(vector1, vector2.T))
```
