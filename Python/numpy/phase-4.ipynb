```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([1,2,3,4,5])
array2 = np.random.random((3,3)) # array2 = np.random.rand(3,3) # without tuple
array3 = np.zeros((4,4))

print(array1)
print(array2)
print(array3)
```

```python
np.save("array1.npy", array1)
np.save("array2.npy", array2)
np.save("array3.npy", array3)
```

```python
loaded_array1 = np.load("array1.npy")
loaded_array2 = np.load("array2.npy")
loaded_array3 = np.load("array3.npy")

print(loaded_array1)
print(loaded_array2)
print(loaded_array3)
```

```python

# print(loaded_numpy_logo)

try:
    loaded_numpy_logo = np.load("numpy-logo.npy")
    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(loaded_numpy_logo)
    plt.title("NumPy Logo")
    
    dark_log = 1 - loaded_numpy_logo 
    plt.subplot(122)
    plt.imshow(dark_log)
    plt.title("Dark logo")
except FileNotFoundError:
    print("Logo file not found")
```
