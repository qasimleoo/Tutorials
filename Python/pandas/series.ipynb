### Pandas Series 

```python
import pandas as pd
import numpy as np 
print(pd.__version__)
print(np.__version__)
```

### Create series

```python
sales = pd.Series([5,1,6, None , 12]) # list
# sales = pd.Series((5,1,3,5,6)) # tuple
# sales = pd.Series(np.array([5.4,13,3,5,6]))
sales
```

### Indexing and slicing

```python
# sales[1]
sales[1:]
```

### Attributes

```python
print(sales.index)
print(sales.values)
print(sales.dtype)
```

### Methods

```python
# head() - to get the list of specific rows from start
# print(sales.head()) # default to 5
print("head method: ", sales.head(4)) # first 4 rows

# tail() # default to 5
print("\ntale method:", sales.tail(3)) # last 3 rows

# sort()
print("\nsort (sort_values) method:", sales.sort_values())

# count()
print("\ncount method:", sales.count()) # doesn't include null values

# isnull()
print("\nisnull method:", sales.isnull())
```

### Operations

```python
sales1 = pd.Series([1,3,4,5,2])
sales2 = pd.Series([5,1,2,4,5])

# Addition (series)
print("Addition of 2 Series:", sales1 + sales2)

# Addition (scaler) - broadcasting
print("\nAddition of a scaler in Series:", sales2 + 5)

# Multiplication (scaler)
print("\nMultiplication of a scaler in Series:", sales2 * 5)

# Addition (scaler)
print("\nSquare of a scaler in Series:", sales2 ** 5)
```

### Missing values

```python
sales_with_missing = pd.Series([None,2,3,4, None])
print("isnull: ", sales_with_missing.isnull())
print("\nListing:", sales_with_missing)

# fill None/Null with 0 
sales_with_missing.fillna(0)

# .. above is just a view let's store it in the same varibale
sales_with_missing = sales_with_missing.fillna(0)
print("\nFilled with 0s: ", sales_with_missing)
```
