## DataFrame

```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "Name": ["Ali", "Ehmad"],
    "Lemonades": [1, 4],  
})
data
```

```python
s1 = pd.Series(data=[1,2,3,4,5])
s2 = pd.Series(data=[10,20,30,40,50])

# si = pd.Series(data=[100,200,300,400,500], index=list('abcde'))

# data = pd.DataFrame({'C1': s1, 'C2': s2, 'C3': si})
data = pd.DataFrame({'C1': s1, 'C2': s2})
data
```

### Attributes

```python
print(s1.shape)
print(data.shape)

print(data.columns)
print(data.dtypes)
```

### Accessing Data

```python
# Using labels

# loc
# data.loc['a']
print(data.loc[1])  
print(data.loc[1, 'C1']) # you can mention column name here

# iloc
print(data.iloc[1])
print(data.iloc[2, 1])  # you can mention column's index here
```

### Methods

```python
# head()
# tail()
# count()

# info()
data.info() # details
```

```python
# describe()
data.describe() # describe
```

```python
# mean()
data.mean()
```

### Filtering

```python
# print(data['C1'] > 3)
data[data['C1'] > 3]
```

### Adding, Updating, Deleting

```python
# Adding

data['Tips'] = [1,2,3,4, None]
data
```

```python
# Deleting

# del
del data["C1"]
data

# drop
data = data.drop(columns='Tips')
data
```

```python
# Update

# print(data['C1'] == 10)

data.loc[data['C1'] == 3] = 10
data 
```

```python
# Rename

# data.rename(columns={'C1': 'Column1'})  # Temp

data.rename(columns={'C1': 'Column1'}, inplace=True)
data
```

# Multi Index

```python
# multi_index = pd.DataFrame({
#     'Alphas': ['a', 'b', 'c', 'd', 'e'],
#     "Ints": [6,7,8,9,10], 
#     'Floats': [1.25, 2.25, 5.2, 1, 4]
# })
# multi_index.set_index(['Ints', 'Alphas'], inplace=True)

multi_index = pd.DataFrame({
    'Alphas': ['a', 'b', 'c', 'd', 'e'],
    "Ints": [6,7,8,9,10], 
    'Floats': [1.25, 2.25, 5.2, 1, 4]
}).set_index(['Ints', 'Alphas'])
multi_index
```

### Selecting with conditions

```python
value = 3
print(multi_index['Floats'] > value)

print(multi_index[multi_index['Floats'] > value])
```

### Query method 
#### does the same thing without calling the var twice

```python
multi_index.query('Floats > 2')
```
